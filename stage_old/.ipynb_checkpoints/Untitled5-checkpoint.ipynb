```python
import numpy as np
from scipy.optimize import minimize

def PCD(x, y):
    """Calcule la distance basée sur le coefficient de corrélation de Pearson."""
    return 1 - np.corrcoef(x, y)[0, 1]

def loss_function(w, X):
    """Calcule la perte en utilisant la distance PCD."""
    n = len(X)
    return np.sum([PCD(w, xi) for xi in X]) / n

def K_PCD(X):
    """
    Implémente l'algorithme K-PCD pour calculer le centroïde en minimisant PCD.
    
    Paramètres :
        X : liste ou tableau numpy contenant les séries temporelles.

    Retourne :
        C : centroïde optimal trouvé.
    """
    w_guess = np.mean(X, axis=0)  # Estimation initiale avec la moyenne arithmétique
    result = minimize(loss_function, w_guess, args=(X,), method='L-BFGS-B')
    return result.x  # Retourne le centroïde optimal

# Exemple d'utilisation
X = np.random.rand(10, 50)  # 10 séries temporelles de longueur 50
C = K_PCD(X)
print("Centroïde optimisé :", C)
```

```python
import numpy as np
from scipy.optimize import minimize

def euclidean_distance(x, y):
    """Calcule la distance euclidienne entre deux vecteurs."""
    return np.linalg.norm(x - y)

def loss_function_2(params, C_prime, X):
    """Calcule la perte (Loss2) en utilisant l'équation 10."""
    s, t = params
    C_j = s * C_prime + t  # Equation 9 : Transformation du centroïde
    n = len(X)
    return np.sum([euclidean_distance(C_j, xi) for xi in X]) / n

def adjust_centroid(X, C_prime):
    """
    Implémente l'algorithme 2 pour ajuster le centroïde en minimisant la perte.

    Paramètres :
        X : Liste ou tableau numpy des points de données.
        C_prime : Centroïde initial issu de l'algorithme 1.

    Retourne :
        C_j : Centroïde ajusté optimal.
    """
    # Initialisation des paramètres (s, t)
    s_guess = 1.0
    t_guess = 0.0

    # Minimisation de la perte
    result = minimize(loss_function_2, (s_guess, t_guess), args=(C_prime, X), method='L-BFGS-B')

    # Extraction des paramètres optimaux
    s_opt, t_opt = result.x
    C_j = s_opt * C_prime + t_opt
    return C_j

# Exemple d'utilisation
X = np.random.rand(10, 50)  # 10 séries temporelles de longueur 50
C_prime = np.mean(X, axis=0)  # Centroïde initial
C_j = adjust_centroid(X, C_prime)
print("Centroïde ajusté :", C_j)
```

```python
import numpy as np

def detect_low_std_indices(X):
    """
    Implémente l'algorithme 3 pour détecter les indices des séries temporelles avec faible variance.

    Paramètres :
        X : Tableau numpy représentant les séries temporelles (n_samples, n_features).

    Retourne :
        low_std_indices : Indices des séries temporelles ayant peu de variation.
    """
    # Étape 1 : Trouver les séries où 90 % ou plus des valeurs sont identiques
    threshold = 0.9 * X.shape[1]
    L1 = np.where((X == X[:, 0][:, np.newaxis]).sum(axis=1) >= threshold)[0]

    # Étape 2 : Identifier les séries où la différence max-min est ≤ 0.5
    L2 = np.where(np.max(X, axis=1) - np.min(X, axis=1) <= 0.5)[0]

    # Étape 3 : Combiner les indices des deux conditions
    low_std_indices = np.concatenate((L1, L2))

    # Étape 4 : Retourner les indices
    return np.unique(low_std_indices)

# Exemple d'utilisation
X = np.random.rand(10, 50)  # 10 séries temporelles de longueur 50
low_std_indices = detect_low_std_indices(X)
print("Indices avec faible variance :", low_std_indices)
```

```python
import numpy as np
from scipy.optimize import minimize
from sklearn.cluster import KMeans

def PCD_Centroid(X):
    def loss(w):
        return np.mean(np.sum((X - w)**2, axis=1))
    w_guess = np.mean(X, axis=0)
    result = minimize(loss, w_guess)
    return result.x

def update_centroids(X, centroids):
    updated_centroids = []
    for c in centroids:
        def loss(st):
            s, t = st
            adjusted = s * c + t
            return np.mean(np.sum((X - adjusted) ** 2, axis=1))
        result = minimize(loss, [1, 0])
        s, t = result.x
        updated_centroids.append(s * c + t)
    return np.array(updated_centroids)


def detect_low_std_indices(X):
    threshold = 0.9 * X.shape[1]
    
    if isinstance(X, pd.DataFrame):
        L1 = np.where((X.values == X.iloc[:, 0].values[:, np.newaxis]).sum(axis=1) >= threshold)[0]
        L2 = np.where(np.max(X.values, axis=1) - np.min(X.values, axis=1) <= 0.5)[0]
    else:
        L1 = np.where((X == X[:, 0][:, np.newaxis]).sum(axis=1) >= threshold)[0]
        L2 = np.where(np.max(X, axis=1) - np.min(X, axis=1) <= 0.5)[0]
    
    low_std_indices = np.concatenate((L1, L2))
    return np.unique(low_std_indices)

# Algorithme 4 : XK-PCD Clustering principal
def XK_PCD_Clustering(X, k):
    # Étape 1 : Identifier les séries à faible variance
    low_std_indices = detect_low_std_indices(X)

    # Étape 2 : Gérer les séries à faible variance
    low_std_centroid = None
    if len(low_std_indices) > 0:
        low_std_centroid = np.mean(X[low_std_indices], axis=0)
        X = np.delete(X, low_std_indices, axis=0)
        k -= 1

    # Étape 3 : Initialiser K-means
    kmeans = KMeans(n_clusters=k, n_init=10, max_iter=300)
    labels = kmeans.fit_predict(X)
    centroids = kmeans.cluster_centers_

    # Étape 4 : Optimisation avec PCD
    for _ in range(100):
        new_centroids = np.array([PCD_Centroid(X[labels == i]) for i in range(k)])
        changes = np.abs(np.linalg.norm(centroids - new_centroids, axis=1))
        
        if np.all(changes < 0.001):
            break

        centroids = new_centroids

        # Réassigner les points au cluster le plus proche
        labels = np.argmin(np.array([[np.linalg.norm(x - c) for c in centroids] for x in X]), axis=1)

    # Étape 5 : Mise à jour finale des centroïdes
    centroids = update_centroids(X, centroids)

    # Étape 6 : Réintégrer le cluster faible variance si présent
    if low_std_centroid is not None:
        centroids = np.vstack((centroids, low_std_centroid))
        low_std_label = k
        labels = np.append(labels, np.full(len(low_std_indices), low_std_label))

    return centroids, labels

# Exemple d'exécution
X = np.random.rand(100, 10)  # 100 séries temporelles de longueur 10
k = 5  # Nombre de clusters
centroids, labels = XK_PCD_Clustering(X, k)

print("Centroids :\n", centroids)
print("Labels :\n", labels)
```
