```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
from tslearn.preprocessing import TimeSeriesScalerMeanVariance
from tslearn.clustering import TimeSeriesKMeans
from sklearn.decomposition import PCA
```

```python

df_temp_out = pd.read_csv('outdoor_temperature.csv')  
df_office = pd.read_csv('office_consumption.csv')
df_temp_in= pd.read_csv('indoor_temperature.csv')  
df_heat = pd.read_csv('heating_consumption.csv')  
df_cool = pd.read_csv('cooling_consumption.csv')  

```

```python
if df_office.shape[0] == df_heat.shape[0] == df_cool.shape[0]:
    # Créer un DataFrame vide pour les résultats
    df_consumption = pd.DataFrame()

    # Calculer la moyenne pour chaque colonne correspondante
    for col_office, col_heat, col_cool in zip(df_office.columns, df_heat.columns, df_cool.columns):
        # Calculer la moyenne des trois colonnes pour chaque ligne
        df_consumption[col_office] = (df_office[col_office] + df_heat[col_heat] + df_cool[col_cool]) 

    # Afficher le DataFrame final avec les moyennes
    print(df_consumption)
else:
    print("Les DataFrames n'ont pas le même nombre de lignes.")
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
df_consumption_scaled = pd.DataFrame(scaler.fit_transform(df_consumption), columns=df_consumption.columns)
```

```python

kmeans_dtw = TimeSeriesKMeans(n_clusters=4, metric="dtw", random_state=42)
df_consumption_scaled['cluster'] = kmeans_dtw.fit_predict(df_consumption_scaled)

```

```python
fig, axes = plt.subplots(2, 2, figsize=(14, 12))  

for cluster_id in range(kmeans_dtw.n_clusters):
    # Sélectionner les données appartenant au cluster actuel
    cluster_data = df_consumption_scaled[df_consumption_scaled['cluster'] == cluster_id].drop(columns=['cluster'])

    for i in range(len(cluster_data)):
        axes[cluster_id // 2, cluster_id % 2].plot(range(24), cluster_data.iloc[i], color='gray', alpha=0.5)

    # Tracer le centre du cluster
    axes[cluster_id // 2, cluster_id % 2].plot(range(24), kmeans_dtw.cluster_centers_[cluster_id], color='red', linewidth=2, label=f"Centre Cluster {cluster_id}")

    axes[cluster_id // 2, cluster_id % 2].set_title(f"Cluster {cluster_id}")
    axes[cluster_id // 2, cluster_id % 2].set_xlabel("Heure (0 à 23)")
    axes[cluster_id // 2, cluster_id % 2].set_ylabel("Valeur")
    axes[cluster_id // 2, cluster_id % 2].grid(True)
    axes[cluster_id // 2, cluster_id % 2].legend()

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np

# Charger les fichiers CSV
df_temp_out = pd.read_csv('outdoor_temperature.csv')  
df_office = pd.read_csv('office_consumption.csv')
df_temp_in = pd.read_csv('indoor_temperature.csv')  
df_heat = pd.read_csv('heating_consumption.csv')  
df_cool = pd.read_csv('cooling_consumption.csv')  

# Réinitialiser les noms des colonnes si elles sont incorrectes
# Supposons que tu veuilles simplement remplacer les noms par des heures (Heure_1, Heure_2, ...)
num_columns = df_temp_out.shape[1]  # Nombre de colonnes (en supposant qu'ils ont tous 24 colonnes)
new_column_names = [f"Heure_{i+1}" for i in range(num_columns)]  # Créer une liste de noms comme "Heure_1", "Heure_2", ...

# Appliquer ces nouveaux noms à chaque DataFrame
df_temp_out.columns = new_column_names
df_office.columns = new_column_names
df_temp_in.columns = new_column_names
df_heat.columns = new_column_names
df_cool.columns = new_column_names


df_combined = pd.DataFrame()

# Pour chaque colonne (représentant chaque heure)
for col in df_temp_out.columns:
    # Créer un vecteur pour chaque ligne avec les valeurs correspondantes de chaque CSV
    df_combined[col] = [
        list(zip(
            [df_temp_out[col][i]],  # Chaque valeur est maintenant un tuple ou une liste
            [df_office[col][i]],
            [df_temp_in[col][i]],
            [df_heat[col][i]],
            [df_cool[col][i]]
        )) 
        for i in range(len(df_temp_out))
    ]

# Sauvegarder le DataFrame combiné
df_combined.to_csv('combined_data.csv', index=False)

# Afficher un aperçu du DataFrame combiné
print(df_combined.head())
```

```python
# Créer un DataFrame vide pour les résultats des médianes
df_median = pd.DataFrame()

# Calculer la médiane des vecteurs pour chaque cellule de df_combined
for col in df_combined.columns:
    # Appliquer np.median sur chaque vecteur pour obtenir une seule valeur
    df_median[col] = df_combined[col].apply(lambda x: np.median(np.array(x), axis=0)[0])

# Sauvegarder le DataFrame avec les médianes

# Afficher un aperçu du DataFrame avec les médianes
print(df_median.head())
```

```python

scaler = TimeSeriesScalerMeanVariance()
data_normalized3 = scaler.fit_transform(df_median.values.reshape((df_median.shape[0], df_median.shape[1], 1)))  
data_scaled = pd.DataFrame(data_normalized3.reshape(data_normalized3.shape[0], data_normalized3.shape[1]), columns=df_median.columns)
print( data_scaled)
```

```python

kmeans_dtw = TimeSeriesKMeans(n_clusters=4, metric="dtw", random_state=42)
df_median['cluster'] = kmeans_dtw.fit_predict(df_median)

```

```python
fig, axes = plt.subplots(2, 2, figsize=(14, 12))

for cluster_id in range(kmeans_dtw.n_clusters):
    cluster_data = df_median[df_median['cluster'] == cluster_id]

    for idx, row in cluster_data.iterrows():
        axes[cluster_id // 2, cluster_id % 2].plot(range(24), row[:-1], color='gray', alpha=0.5)
    
    cluster_center = kmeans_dtw.cluster_centers_[cluster_id]
    
    axes[cluster_id // 2, cluster_id % 2].plot(range(24), cluster_center.flatten(), color='red', linewidth=3, label=f"Centre du cluster {cluster_id}")
    
    axes[cluster_id // 2, cluster_id % 2].set_title(f"Séries temporelles du Cluster {cluster_id}")
    axes[cluster_id // 2, cluster_id % 2].set_xlabel("Heures")
    axes[cluster_id // 2, cluster_id % 2].set_ylabel("valeur")
    axes[cluster_id // 2, cluster_id % 2].grid(True)

axes[1, 1]

plt.tight_layout()
plt.show()
```
