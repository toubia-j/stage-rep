```python
from tslearn.preprocessing import TimeSeriesScalerMeanVariance
import numpy as np
from scipy.optimize import minimize
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import pandas as pd
from tslearn.clustering import TimeSeriesKMeans
```

```python

def PCD_Centroid(X):
    def loss(w):
        return np.mean(np.sum((X - w)**2, axis=1))
    w_guess = np.mean(X, axis=0)
    result = minimize(loss, w_guess)
    return result.x

def update_centroids(X, centroids, labels):
    updated_centroids = []
    for i, c in enumerate(centroids):
        cluster_points = X[labels == i]
        if len(cluster_points) == 0:  
            updated_centroids.append(c)
            continue
        
        def loss(st):
            s, t = st
            adjusted = s * c + t
            return np.mean(np.sum((cluster_points - adjusted) ** 2, axis=1))

        result = minimize(loss, [1, 0])
        s, t = result.x
        updated_centroids.append(s * c + t)

    return np.array(updated_centroids)


def detect_low_std_indices(X):
    threshold =  0.9 *X.shape[1]
    
    if isinstance(X, pd.DataFrame):
        L1 = np.where((X.values == X.iloc[:, 0].values[:, np.newaxis]).sum(axis=1) >= threshold)[0]
        #L2 = np.where(np.max(X.values, axis=1) - np.min(X.values, axis=1) <= 0.5)[0]
    else:
        L1 = np.where((X == X[:, 0][:, np.newaxis]).sum(axis=1) >= threshold)[0]
       # L2 = np.where(np.max(X, axis=1) - np.min(X, axis=1) <= 0.5)[0]
    low_std_indices=L1
    #low_std_indices = np.concatenate((L1, L2))
    return np.unique(low_std_indices)

def XK_PCD_Clustering(X, k):
    """
    low_std_indices = detect_low_std_indices(X)

    # Étape 2 : Gérer les séries à faible variance
    low_std_centroid = None
    if len(low_std_indices) > 0:
        low_std_centroid = np.mean(X.iloc[low_std_indices], axis=0)
        
        if isinstance(X, pd.DataFrame):
            # On crée une copie du DataFrame en supprimant les indices de faibles variances
            X = X.drop(X.index[low_std_indices])
        else:
            # On supprime les lignes dans ndarray en s'assurant de l'intégrité des indices
            X = np.delete(X, low_std_indices, axis=0)
        
        k -= 1
    """
    kmeans = KMeans(n_clusters=k, n_init=20, max_iter=300 )
    labels = kmeans.fit_predict(X)
    centroids = kmeans.cluster_centers_

    for _ in range(100):
        new_centroids = np.array([PCD_Centroid(X[labels == i]) for i in range(k)])
        changes = np.abs(np.linalg.norm(centroids - new_centroids, axis=1))
        
        if np.all(changes < 0.001):
            break
        
        centroids = new_centroids
    
        labels = np.argmin(np.array([[np.linalg.norm(x - c) for c in centroids] for x in X]), axis=1)
    
    

    centroids = update_centroids(X, centroids,labels)
    """    
    if low_std_centroid is not None:
        centroids = np.vstack((centroids, low_std_centroid))
        low_std_label = k
        labels = np.append(labels, np.full(len(low_std_indices), low_std_label))
    """
   

    return centroids, labels
```

```python
def normalize_and_scale_timeseries(df_median):
    scaler = TimeSeriesScalerMeanVariance()
    data_normalized3 = scaler.fit_transform(df_median.values.reshape((df_median.shape[0], df_median.shape[1], 1)))
    data_scaled = pd.DataFrame(data_normalized3.reshape(data_normalized3.shape[0], data_normalized3.shape[1]), columns=df_median.columns)
    return data_scaled
```

```python

def plot_cluster_timeseries(data_scaled2, labels, centroids,nbcluster, ylabel="Valeur"):
    data_scaled2['cluster'] = pd.Series(labels, index=data_scaled2.index)
    fig, axes = plt.subplots(2, 2, figsize=(14, 12))

    for cluster_id in range(nbcluster):
        cluster_data = data_scaled2[data_scaled2['cluster'] == cluster_id]

        for idx, row in cluster_data.iterrows():
            axes[cluster_id // 2, cluster_id % 2].plot(range(24), row[:-1], color='gray', alpha=0.5)

        cluster_center = centroids[cluster_id]
        axes[cluster_id // 2, cluster_id % 2].plot(range(24), cluster_center.flatten(), color='red', linewidth=3, label=f"Centre du cluster {cluster_id}")

        cluster_median = cluster_data.iloc[:, :-1].mean(axis=0)
        axes[cluster_id // 2, cluster_id % 2].plot(range(24), cluster_median, color='blue', linewidth=2, linestyle='dotted', label=f"Médiane du cluster {cluster_id}")

        axes[cluster_id // 2, cluster_id % 2].set_title(f"Séries temporelles du Cluster {cluster_id}")
        axes[cluster_id // 2, cluster_id % 2].set_xlabel("Heures")
        axes[cluster_id // 2, cluster_id % 2].set_ylabel(ylabel)
        axes[cluster_id // 2, cluster_id % 2].grid(True)
    axes[1, 1].axis('off')

    plt.tight_layout()
    plt.show()
```

```python
df_temp_out = pd.read_csv('outdoor_temperature.csv')  

#data_scaled= normalize_and_scale_timeseries(df_temp_out)
centroids, labels = XK_PCD_Clustering(df_temp_out, 3)
plot_cluster_timeseries(df_temp_out,labels,centroids,3,"outdoor temperature")
#print("Centroids :\n", centroids)
#print("Labels :\n", labels)
```
