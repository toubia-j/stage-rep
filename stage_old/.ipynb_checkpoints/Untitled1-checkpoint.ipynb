```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from sklearn.metrics import silhouette_samples, silhouette_score
from tslearn.preprocessing import TimeSeriesScalerMeanVariance
from tslearn.clustering import TimeSeriesKMeans
from tslearn.clustering import KShape
```

```python
df = pd.read_csv("outdoor_temperature.csv")
print(df.head())
data=df.copy()
```

```python
scaler = StandardScaler()
data_normalized = scaler.fit_transform(data)
```

```python
inertias = []
range_n_clusters = range(1, 11)

for n_clusters in range_n_clusters:
    kmeans = KMeans(n_clusters=n_clusters, random_state=42)
    kmeans.fit(data_normalized)
    inertias.append(kmeans.inertia_)

# Tracer le graphique
plt.plot(range_n_clusters, inertias, marker='o')
plt.title("Méthode du Coude pour le nombre optimal de clusters")
plt.xlabel("Nombre de clusters")
plt.ylabel("Inertie")
plt.show()
```

```python
silhouette_scores = []
for n_clusters in range(2, 5):
    kmeans = KMeans(n_clusters=n_clusters, random_state=42)
    cluster_labels = kmeans.fit_predict(data_normalized)
    
    # Calculer le score de silhouette global
    score = silhouette_score(data_normalized, cluster_labels)
    silhouette_scores.append(score)
    
    # Afficher les scores de silhouette pour chaque point
    sample_silhouette_values = silhouette_samples(data_normalized, cluster_labels)
    
    # Tracer le score de silhouette pour chaque point
    plt.figure(figsize=(7, 5))
    y_lower = 10
    for i in range(n_clusters):
        cluster_silhouette_values = sample_silhouette_values[cluster_labels == i]
        cluster_silhouette_values.sort()
        
        size_cluster_i = cluster_silhouette_values.shape[0]
        y_upper = y_lower + size_cluster_i
        
        plt.fill_betweenx(np.arange(y_lower, y_upper), 0, cluster_silhouette_values, alpha=0.7)
        plt.text(-0.05, (y_lower + y_upper) / 2, f'Cluster {i + 1}', fontsize=12)
        
        y_lower = y_upper + 10  # Ajout d'un petit espace entre les clusters
    
    plt.title(f"Silhouette plot for n_clusters = {n_clusters}")
    plt.xlabel("Silhouette coefficient values")
    plt.ylabel("Cluster labels")
    plt.axvline(x=0, color="gray", linestyle="--")
    plt.show()
```

```python
data_kmeans = data.copy()

```

```python

kmeans = KMeans(n_clusters=3, random_state=42)
data_kmeans['cluster'] = kmeans.fit_predict(data_normalized)
```

```python
print(data_kmeans.head())

```

```python
fig, axes = plt.subplots(2, 2, figsize=(14, 12))

for cluster_id in range(kmeans.n_clusters):
    cluster_data = data_kmeans[data_kmeans['cluster'] == cluster_id]

    for idx, row in cluster_data.iterrows():
        axes[cluster_id // 2, cluster_id % 2].plot(range(24), row[:-1], color='gray', alpha=0.5)
    
    cluster_center = kmeans.cluster_centers_[cluster_id]
    
    cluster_center_denormalized = scaler.inverse_transform([cluster_center])[0]
    
    axes[cluster_id // 2, cluster_id % 2].plot(range(24), cluster_center_denormalized, color='red', linewidth=3, label=f"Centre du cluster {cluster_id}")
    
    axes[cluster_id // 2, cluster_id % 2].set_title(f"Séries temporelles du Cluster {cluster_id}")
    axes[cluster_id // 2, cluster_id % 2].set_xlabel("Heures")
    axes[cluster_id // 2, cluster_id % 2].set_ylabel("Outdoor Temperature")
    axes[cluster_id // 2, cluster_id % 2].grid(True)

axes[1, 1].axis('off')

plt.tight_layout()
plt.show()
```

```python
scaler = TimeSeriesScalerMeanVariance()
data_normalized3 = scaler.fit_transform(data.values.reshape((data.shape[0], data.shape[1], 1)))  
data_normalized2 = pd.DataFrame(data_normalized3.reshape(data_normalized3.shape[0], data_normalized3.shape[1]), columns=data.columns)
```

```python
data_kmeans_dtw = data.copy()

kmeans_dtw = TimeSeriesKMeans(n_clusters=3, metric="dtw", random_state=42)
data_kmeans_dtw['cluster'] = kmeans_dtw.fit_predict(data_normalized2)
data_normalized2['cluster'] = kmeans_dtw.fit_predict(data_normalized2)

```

```python
fig, axes = plt.subplots(2, 2, figsize=(14, 12))

for cluster_id in range(kmeans_dtw.n_clusters):
    cluster_data = data_normalized2[data_normalized2['cluster'] == cluster_id]

    for idx, row in cluster_data.iterrows():
        axes[cluster_id // 2, cluster_id % 2].plot(range(24), row[:-1], color='gray', alpha=0.5)
    
    cluster_center = kmeans_dtw.cluster_centers_[cluster_id]
    
    axes[cluster_id // 2, cluster_id % 2].plot(range(24), cluster_center.flatten(), color='red', linewidth=3, label=f"Centre du cluster {cluster_id}")
    
    axes[cluster_id // 2, cluster_id % 2].set_title(f"Séries temporelles du Cluster {cluster_id}")
    axes[cluster_id // 2, cluster_id % 2].set_xlabel("Heures")
    axes[cluster_id // 2, cluster_id % 2].set_ylabel("Outdoor Temperature (Normalisé)")
    axes[cluster_id // 2, cluster_id % 2].grid(True)

axes[1, 1].axis('off')

plt.tight_layout()
plt.show()
```

```python
scaler = TimeSeriesScalerMeanVariance()
data_normalized3 = scaler.fit_transform(data.values.reshape((data.shape[0], data.shape[1], 1)))  
data_normalized2 = pd.DataFrame(data_normalized3.reshape(data_normalized3.shape[0], data_normalized3.shape[1]), columns=data.columns)
```

```python
data_kshape = data.copy()
kshape = KShape(n_clusters=3, random_state=42)
data_kshape['cluster'] = kshape.fit_predict(data_normalized2.values)
data_normalized2['cluster'] = data_kshape['cluster']
```

```python

fig, axes = plt.subplots(2, 2, figsize=(14, 12))

for cluster_id in range(kshape.n_clusters):
    cluster_data = data_normalized2[data_normalized2['cluster'] == cluster_id]

    for idx, row in cluster_data.iterrows():
        axes[cluster_id // 2, cluster_id % 2].plot(range(24), row[:-1], color='gray', alpha=0.5)
    
    cluster_center = kshape.cluster_centers_[cluster_id]
    
    axes[cluster_id // 2, cluster_id % 2].plot(range(24), cluster_center, color='red', linewidth=3, label=f"Centre du cluster {cluster_id}")
    
    axes[cluster_id // 2, cluster_id % 2].set_title(f"Séries temporelles du Cluster {cluster_id}")
    axes[cluster_id // 2, cluster_id % 2].set_xlabel("Heures")
    axes[cluster_id // 2, cluster_id % 2].set_ylabel("Outdoor Temperature (Normalisé)")
    axes[cluster_id // 2, cluster_id % 2].grid(True)

axes[1, 1].axis('off')

plt.tight_layout()
plt.show()
```
