```python
import pandas as pd
import numpy as np
from tslearn.preprocessing import TimeSeriesScalerMeanVariance
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from sklearn.metrics import silhouette_score
from sklearn.metrics import davies_bouldin_score
```

```python
df_temp_out = pd.read_csv('outdoor_temperature.csv')  
df_temp_in = pd.read_csv('indoor_temperature.csv')  
df_heat = pd.read_csv('heating_consumption.csv')  

df_combined_tempout = pd.concat([df_temp_out ,df_heat], axis=1)  
df_combined_tempin = pd.concat([ df_temp_in, df_heat], axis=1)  

date_range = pd.date_range(start='2021-01-01', end='2022-12-28', freq='D')

df_combined_tempout['date'] = date_range
df_combined_tempin['date'] = date_range

print(df_combined_tempout)
```

```python
df_no_heating = df_combined_tempout[df_combined_tempout.iloc[:, -25:-1].sum(axis=1) == 0]  
df_heating_tempout = df_combined_tempout[df_combined_tempout.iloc[:, -25:-1].sum(axis=1) != 0]      

df_heating_tempout = df_heating_tempout.reset_index(drop=True)

#print(df_heating_tempout)
df_heating2_tempout = df_heating_tempout.drop(df_heating_tempout.columns[-25:], axis=1)

df_no_heating2 = df_combined_tempin[df_combined_tempin.iloc[:, -25:-1].sum(axis=1) == 0]  
df_heating_tempin = df_combined_tempin[df_combined_tempin.iloc[:, -25:-1].sum(axis=1) != 0]      

df_heating_tempin = df_heating_tempin.reset_index(drop=True)

#print(df_heating_tempin)
df_heating2_tempin = df_heating_tempin.drop(df_heating_tempin.columns[-25:], axis=1)
```

```python

def apply_kmeans(data, n_clusters=3):
    kmeans = KMeans(n_clusters=n_clusters, random_state=42)
    data['cluster'] = kmeans.fit_predict(data)
    
    return kmeans, data
```

```python

def plot_clusters_single_variable(data_normalized, cluster_assignments, kmeans, column_index, y_label="Valeur"):
    fig, axes = plt.subplots(2, 2, figsize=(14, 12))

    all_values = data_normalized.iloc[:, column_index:].values.flatten()
    y_min, y_max = np.min(all_values), np.max(all_values)

    for cluster_id in range(kmeans.n_clusters):
        cluster_data = data_normalized[cluster_assignments == cluster_id].iloc[:, column_index:]
        cluster_center = kmeans.cluster_centers_[cluster_id, column_index:-1]

        for _, row in cluster_data.iterrows():
            axes[cluster_id // 2, cluster_id % 2].plot(range(24), row, color='gray', alpha=0.5)

        axes[cluster_id // 2, cluster_id % 2].plot(range(24), cluster_center, color='red', linewidth=3, label=f"Centre du cluster {cluster_id}")
        
        axes[cluster_id // 2, cluster_id % 2].set_title(f"SÃ©ries temporelles du Cluster {cluster_id}")
        axes[cluster_id // 2, cluster_id % 2].set_xlabel("Heures")
        axes[cluster_id // 2, cluster_id % 2].set_ylabel(y_label)
        axes[cluster_id // 2, cluster_id % 2].set_ylim([y_min, y_max])
        axes[cluster_id // 2, cluster_id % 2].grid(True)

    plt.tight_layout()
    plt.show()

```

```python
print(df_heating_tempout)
```

```python

kmeans, data_tempout_with_cluster = apply_kmeans(df_heating2_tempout, 4)

df_heating_tempout["cluster"]=data_tempout_with_cluster["cluster"]
```

```python
kmeans, data_tempin_with_cluster = apply_kmeans(df_heating2_tempin, 4)

df_heating_tempin["cluster"]=data_tempin_with_cluster["cluster"]
```

```python
df_heating2_tempout2=df_heating2_tempout.drop(columns=["cluster"])
plot_clusters_single_variable(df_heating2_tempout2, df_heating2_tempout["cluster"],kmeans, column_index=0, y_label="outdoor_temperature")
```

```python

labels = df_heating2_tempout['cluster'] 
df_heating2_tempout=df_heating2_tempout.drop(columns=["cluster"])

sil_score = silhouette_score(df_heating2_tempout, labels, metric='euclidean')
db_score = davies_bouldin_score(df_heating2_tempout, labels)

print(f"Davies-Bouldin Index: {db_score}")
print(f"Silhouette Score: {sil_score}")
```

```python
df_heating2_tempin2=df_heating2_tempin.drop(columns=["cluster"])

plot_clusters_single_variable(df_heating2_tempin2, df_heating2_tempin["cluster"],kmeans, column_index=0, y_label="indoor_temperature")
```
