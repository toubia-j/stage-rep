```python
import pandas as pd
import numpy as np
from tslearn.preprocessing import TimeSeriesScalerMeanVariance
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from sklearn.metrics import silhouette_score
from sklearn.metrics import davies_bouldin_score
```

```python
import pandas as pd

df = pd.read_csv("heating_consumption.csv", header=None)

print(df.head())
```

```python
df_heat = pd.read_csv('heating_consumption.csv', header=None)  
date_range = pd.date_range(start='2021-01-01', end='2022-12-29', freq='D')

df_heat['date'] = date_range
```

```python
import holidays

start_date = '2021-01-01'
end_date = '2022-12-29'

fr_holidays = holidays.France(years=[2021, 2022])

holidays_in_range = [day for day in fr_holidays if start_date <= str(day) <= end_date]

additional_holidays = pd.to_datetime(['2021-12-27', '2021-12-28', '2021-12-29', '2021-12-30', '2021-12-31','2021-5-24','2021-11-1','2022-6-6','2022-11-1'])
holidays_in_range.extend(additional_holidays)

print(holidays_in_range)
```

```python
df_heat.insert(0, 'heating_status', (df_heat.iloc[:, -25:-1].sum(axis=1) == 0).astype(int) ^ 1)



df_heat['heating_status'] = df_heat.apply(lambda row: 0 if row['date'].weekday() >= 5 else row['heating_status'], axis=1)

df_heat['heating_status'] = df_heat.apply(
   lambda row: 0 if row['date'].weekday() >= 5 or row['date'] in fr_holidays else row['heating_status'], axis=1)
```

```python
df_heat[df_heat["heating_status"]==0].to_csv("df_heat00.csv", index=False)	
```

```python

def apply_kmeans(data, n_clusters=3):
    kmeans = KMeans(n_clusters=n_clusters, random_state=42)
    df_heating_with_cluster=data.copy()
    df_heating_with_cluster['cluster'] = kmeans.fit_predict(data)
    return kmeans, df_heating_with_cluster
```

```python


def plot_clusters_single_variable(data_normalized, cluster_assignments, column_index=0, y_label="Valeur", num_labels=10):
    cluster_centers = []
    for cluster_id in np.unique(cluster_assignments):
        
        cluster_data = data_normalized[cluster_assignments == cluster_id]
        cluster_center = cluster_data.iloc[:, :-1].mean().values
    
        cluster_centers.append(cluster_center)
    
    cluster_centers = np.array(cluster_centers)
    
    fig, axes = plt.subplots(2, 2, figsize=(14, 12))

    all_values = []
    for cluster_id, cluster_center in enumerate(cluster_centers):
        cluster_data = data_normalized[cluster_assignments == cluster_id].iloc[:, column_index:-1]
      
        all_values.extend(cluster_data.values.flatten()) 
    y_min = np.min(all_values)
    y_max = np.max(all_values)
    
    for cluster_id, cluster_center in enumerate(cluster_centers):
        cluster_data = data_normalized[cluster_assignments == cluster_id].iloc[:, column_index:-1]

        num_series = len(cluster_data)
        
 
        for idx, row in cluster_data.iterrows():
            axes[cluster_id // 2, cluster_id % 2].plot(range(24), row, color='gray', alpha=0.5)

        axes[cluster_id // 2, cluster_id % 2].plot(range(24), cluster_center, color='red', linewidth=3, 
                                                     label=f"{num_series} séries")

        axes[cluster_id // 2, cluster_id % 2].set_title(f"Séries temporelles du Cluster {cluster_id}")
        axes[cluster_id // 2, cluster_id % 2].set_xlabel("Heures")
        axes[cluster_id // 2, cluster_id % 2].set_ylabel(y_label)
        
        axes[cluster_id // 2, cluster_id % 2].set_ylim([y_min, y_max])
        axes[cluster_id // 2, cluster_id % 2].grid(True)

        axes[cluster_id // 2, cluster_id % 2].legend(loc='upper right', bbox_to_anchor=(1.1, 1))

    plt.tight_layout()
    plt.show()
```

```python
def plot_cluster_centers(kmeans, y_label="Valeur"):
    cluster_centers = kmeans.cluster_centers_

    num_clusters = len(cluster_centers)
    colors = plt.cm.get_cmap('tab10', num_clusters) 
    plt.figure(figsize=(10, 6))

    for cluster_id, cluster_center in enumerate(cluster_centers):
        plt.plot(range(24), cluster_center, color=colors(cluster_id), linewidth=3, label=f"Centre du cluster {cluster_id}")

    plt.title("Centres des Clusters")
    plt.xlabel("Heures")
    plt.ylabel(y_label)
    plt.legend(loc='upper right', bbox_to_anchor=(1.1, 1))
    plt.grid(True)
    plt.tight_layout()

    plt.show()
```

```python
def plot_cluster_centers_with_colors(data_normalized, cluster_assignments, y_label="Valeur"):
    cluster_centers = []
    for cluster_id in np.unique(cluster_assignments):
        cluster_data = data_normalized[cluster_assignments == cluster_id]
        cluster_center = cluster_data.iloc[:, :-1].mean().values  # Ignorer la dernière colonne si nécessaire
        cluster_centers.append(cluster_center)

    cluster_centers = np.array(cluster_centers)

    num_clusters = len(cluster_centers)
    colors = plt.cm.get_cmap('tab10', num_clusters)  # Utilisation de la palette 'tab10'

    plt.figure(figsize=(10, 6))

    for cluster_id, cluster_center in enumerate(cluster_centers):
        plt.plot(range(24), cluster_center, color=colors(cluster_id), linewidth=3, label=f"Centre du cluster {cluster_id}")

    plt.title("Centres des Clusters")
    plt.xlabel("Heures")
    plt.ylabel(y_label)
    plt.legend(loc='upper right', bbox_to_anchor=(1.1, 1))
    plt.grid(True)
    plt.tight_layout()

    plt.show()
```

```python
df_heating_status_1 = df_heat[df_heat['heating_status'] == 1]
df_heating_status_0 = df_heat[df_heat['heating_status'] == 0]
df_heating = df_heating_status_1.drop(columns=["date", "heating_status"]) 

kmeans, df_heating_with_cluster_1 = apply_kmeans(df_heating, 3) 


df_heating_with_cluster_1['date'] = df_heating_status_1['date']
df_heating_with_cluster_1['heating_status'] = df_heating_status_1['heating_status']

df_heating_status_0['cluster'] = 3 
df_heating_with_cluster = pd.concat([df_heating_with_cluster_1, df_heating_status_0])

df_heating_with_cluster = df_heating_with_cluster.sort_values(by='date')

df_heating_with_cluster.to_csv("df_heating_conso_with_cluster_01.csv", index=False)
```

```python
df_heating_with_cluster2=df_heating_with_cluster.drop(columns=["heating_status","date"])

plot_clusters_single_variable(df_heating_with_cluster2, df_heating_with_cluster2["cluster"], y_label="heating_consumption ")
```

```python
plot_cluster_centers_with_colors(df_heating_with_cluster2, df_heating_with_cluster2["cluster"], y_label="heating_consumption ")
```

```python
df_temp_out = pd.read_csv('outdoor_temperature.csv', header=None)  
df_temp_in = pd.read_csv('indoor_temperature.csv', header=None)  
df_heat = pd.read_csv('df_heating_conso_with_cluster_01.csv')  
date_range = pd.date_range(start='2021-01-01', end='2022-12-29', freq='D')

df_temp_out.insert(0, 'date', date_range)
df_temp_in.insert(0, 'date', date_range)

temp_out_cols = df_temp_out.columns[1:] 
new_temp_out_cols = {col: f'TEXT{i+1}' for i, col in enumerate(temp_out_cols)}
df_temp_out.rename(columns=new_temp_out_cols, inplace=True)



temp_in_cols = df_temp_in.columns[1:]
new_temp_in_cols = {col: f'TINT{i+1}' for i, col in enumerate(temp_in_cols)}
df_temp_in.rename(columns=new_temp_in_cols, inplace=True)


heat_cols = df_heat.columns[:-3] 
new_heat_cols = {col: f'CONSO{i+1}' for i, col in enumerate(heat_cols)}
df_heat.rename(columns=new_heat_cols, inplace=True)

df_temp_out['date'] = pd.to_datetime(df_temp_out['date'])
df_temp_in['date'] = pd.to_datetime(df_temp_in['date'])
df_heat['date'] = pd.to_datetime(df_heat['date'], errors='coerce')
df_combined = df_heat.merge(df_temp_out, on='date', how='inner').merge(df_temp_in, on='date', how='inner')


df_combined.to_csv("df_final_01.csv", index=False)


```
