```python
#Matrice de corrlation sur les donnes de 4 ans ((jours avec chauffage on seulement )), en comparant la moyenne des 24 heures de chaque jour.

grandeurs = {
    "Occupation": occupation_ouvert,
    "Temperature ext ": temperature_ext_ouvert,
    "Humidity": humidity_ouvert,
    "Temperature int ": temperature_int_ouvert,
    "consommation chauffage":conso_tou_ouvert
}

daily_means = {name: df.mean(axis=1) for name, df in grandeurs.items()}

merged_df = pd.DataFrame(daily_means) 
correlation_matrix = merged_df.corr()


plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".3f")
plt.title("Matrice de corrlation entre les grandeurs")
plt.show()
```

```python
# Matrice de corrlation sur les donnes de 4 ans, en comparant chaque valeur des 24 heures, puis en faisant la moyenne pour obtenir une seule valeur.
correlations = {}
names = list(grandeurs.keys())

for name1 in names:
    for name2 in names:
        if name1 <= name2:  # viter les doublons (matrice symtrique)
            df1, df2 = grandeurs[name1], grandeurs[name2]
            col_corrs = [df1.iloc[:, i].corr(df2.iloc[:, i]) for i in range(df1.shape[1])]
            avg_corr = np.nanmean(col_corrs)
            correlations[(name1, name2)] = avg_corr
            correlations[(name2, name1)] = avg_corr  # pour complter la matrice

# tape 2 : Conversion en DataFrame
correlation_matrix = pd.DataFrame(index=names, columns=names)

for (name1, name2), value in correlations.items():
    correlation_matrix.loc[name1, name2] = value

correlation_matrix = correlation_matrix.astype(float)

plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Matrice de corrlation moyenne horaire entre les grandeurs")
plt.tight_layout()
plt.show()
```
