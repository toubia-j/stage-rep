```python
occupation_new=pd.read_csv("../data/simulation_data/Occupancy_per_hour_Youssef.txt")
occ=occupation_new.iloc[:, -1]
reshaped = occ.values.reshape(-1, 24)
occupation = pd.DataFrame(reshaped)
```

```python
conso_tou = pd.read_csv("../data/simulation_data/conso_kmeans_data/clusters_dataavecoccupation_heat_perif_toulouse_kmeans")
```

```python
"""
Fusion des DataFrames de  Text et occupation  pour la  zone climatique : Toulouse
"""
Text_occ_agen = pd.concat([Text_2_agen,occupation],axis=1).reset_index(drop=True)
Text_occ_albi = pd.concat([Text_2_albi,occupation],axis=1).reset_index(drop=True)
Text_occ_auch = pd.concat([Text_2_auch,occupation],axis=1).reset_index(drop=True)
Text_occ_toulouse = pd.concat([Text_2_toulouse,occupation],axis=1).reset_index(drop=True)
Text_occ_combined_tou = pd.concat([Text_occ_agen, Text_occ_albi,Text_occ_auch,Text_occ_toulouse], axis=0).reset_index(drop=True)

```

```python
"""
Ajout d'une colonne binaire (pour identifier les jours de consommation)  et d'une colonne cluster pour le  DataFrame
de température à l'extérieur et consommation  de la  zone  climatique : toulouse .
"""
Text_occ_conso_heat_tou2=Text_occ_combined_tou.copy()
Text_occ_conso_heat_tou2['heat_on'] = conso_tou['heat_on']
Text_occ_conso_heat_tou2['clusters'] = conso_tou['clusters']
Text_occ_conso_heat_tou2.columns = Text_occ_conso_heat_tou2.columns.astype(str)
```

```python
"""
 LSTM pour Prédiction de la consommation de chauffage sur 24 heures à Toulouse et sa périphérie,
en fonction de ( température extérieure et cluster  et de la consommation de chauffage ) à t-1 
          et  la température extérieure  + cluster de la consommation de chauffage ((( predit avec tous les models))) à t .
Les 80% premiers jours sont utilisés pour l'entraînement, et les 20% restants pour le test.
L'entraînement et les résultats sont réalisés avec le profil de consommation obtenu à partir de chaque modèle individuellement.
"""

Text_occ_conso_heat_tou = Text_occ_conso_heat_tou2.copy()
Text_occ_conso_heat_tou['heat_on'] = conso_tou['heat_on']
Text_occ_conso_heat_tou = pd.concat([Text_occ_conso_heat_tou, conso_tou.iloc[:, :-2]], axis=1).reset_index(drop=True)
Text_occ_conso_heat_tou.columns = Text_occ_conso_heat_tou.columns.astype(str)

df2 = Text_occ_conso_heat_tou.copy()
scaler_temp2 = StandardScaler() 
scaler_cons2 = StandardScaler()

df2 = df2.applymap(lambda x: float(str(x).replace(',', '.')))

df_scaled2 = np.hstack([
    scaler_temp2.fit_transform(df2.iloc[:, :48]),   
    df2.iloc[:, 48:50].values,  # Colonne heat_on 
    scaler_cons2.fit_transform(df2.iloc[:, 50:]),  
])

df_final2 = pd.DataFrame(df_scaled2, columns=df2.columns)
data2 = df_final2.values  
time_steps = 1  

X2, y2 = [], []
for i in range(1, len(data2)):
    X2.append(np.hstack([data2[i-1, :], data2[i, :48], data2[i, 48:49]])) 
    y2.append(data2[i, 50:]) 
    
X2, y2 = np.array(X2), np.array(y2)
print(df2.shape)
print(X2.shape)
print(y2.shape)
clusters_test = pd.read_csv("../data/simulation_data/conso_kmeans_data/resultat_prediction_dataavecoccupation_perif_toulouse_dernier_20%.csv")

X_test2 = X2[int((X2.shape[0] * 8) / 10):, :]

results = []

for model_name in clusters_test.columns:
    if model_name.startswith('y_pred_'): 
        model_name_clean = model_name.replace('y_pred_', '')
        X_test_temp = np.hstack([X_test2, clusters_test[model_name].values.reshape(-1, 1)])

        clusters_train = conso_tou['clusters'].iloc[:int((X2.shape[0] * 8) / 10) ].values.reshape(-1, 1)
        X_train2 = X2[:int((X2.shape[0] * 8) / 10), :]
        X_train2 = np.hstack([X_train2, clusters_train])

        X_train2 = X_train2.reshape(X_train2.shape[0], 1, X_train2.shape[1])
        X_test_temp = X_test_temp.reshape(X_test_temp.shape[0], 1, X_test_temp.shape[1])

        y_train2 = y2[:int((y2.shape[0] * 8) / 10), :]
        y_test2 = y2[int((y2.shape[0] * 8) / 10):, :]

        model2 = Sequential()
        model2.add(LSTM(68, activation='tanh', input_shape=(X_train2.shape[1], X_train2.shape[2])))  
        model2.add(Dropout(0.2))
        model2.add(Dense(24, activation='linear'))   

        history2, loss2, mae2, mse2, rmse2, mae_test2, mse_test2, rmse_test2, r2, cvrmse, predictions2 = train_and_evaluate(
            model2, X_train2, y_train2, X_test_temp, y_test2, scaler_temp2, scaler_cons2)
        
        results.append({
            "model": model_name_clean,
            "loss": loss2,
            "mae": mae2,
            "mse": mse2,
            "rmse": rmse2,
            "mae_test": mae_test2,
            "mse_test": mse_test2,
            "rmse_test": rmse_test2,
            "r2": r2,
            "cvrmse": cvrmse,
        })

for result in results:
    print("*" * 50)
    print("*" * 50)
    print(f"Résultats pour le modèle {result['model']}:")
    print(f"Loss: {result['loss']}, MAE: {result['mae']}, MSE: {result['mse']}, RMSE: {result['rmse']}")
    print(f"MAE Test: {result['mae_test']}, MSE Test: {result['mse_test']}, RMSE Test: {result['rmse_test']}")
    print(f"R2: {result['r2']}, CVRMSE: {result['cvrmse']}")
    print("*" * 50)
    print("*" * 50)
```
