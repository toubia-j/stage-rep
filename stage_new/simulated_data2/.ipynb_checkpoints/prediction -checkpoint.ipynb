```python

"""irina dataset
"""

def prepare_dataframe(df, ville):
    df_copy = df.copy() 
    dates = pd.date_range(start='2024-01-01', periods=len(df_copy), freq='D') 
    df_copy.insert(0, 'ville', ville) 
    df_copy.insert(1, 'date', dates)  
    return df_copy


Text_agen_copy = prepare_dataframe(Text_agen, "Agen")
Text_albi_copy = prepare_dataframe(Text_albi, "Albi")
Text_auch_copy = prepare_dataframe(Text_auch, "Auch")
Text_toulouse_copy = prepare_dataframe(Text_toulouse, "Toulouse")

Text_combined_tou_copy = pd.concat([Text_agen_copy, Text_albi_copy,Text_auch_copy,Text_toulouse_copy], axis=0).reset_index(drop=True)


df_irina = pd.concat([Text_combined_tou_copy, conso_tou], axis=1, ignore_index=True)

new_columns = ['ville', 'date']+['TEXT_' + str(i) for i in range(1, 25)] + ['conso_' + str(i) for i in range(1, 25)] + ['heating_status', 'cluster']


df_irina.columns = new_columns

df_irina.to_csv("df_irina.csv",sep='\t',index=False)
print(df_irina.head())
```
